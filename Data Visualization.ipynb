#LOUVAIN

```python
df_norm = df_normal.transpose()
df_tum = df_tumor.transpose()
```

```python
correlation_matrix_normal = np.corrcoef(df_norm)
correlation_matrix_tumor = np.corrcoef(df_tum)
print("Corr. matrix normal shape: ", correlation_matrix_normal.shape)
print("Corr. matrix tumor shape: ", correlation_matrix_tumor.shape)
```

```python
threshold_edge = 0.9
```

```python
#BUILD EDGE INDEX FOR NORMAL
mask = correlation_matrix_normal > threshold_edge

# Exclude the diagonal elements (assuming the matrix is square)
np.fill_diagonal(mask, False)

# Get the row and column indices where the mask is True
row_indices, col_indices = np.where(mask)

# Create the edge_index tensor
edge_index_normal = np.array([row_indices, col_indices], dtype=np.int64)

print(edge_index_normal.shape)
```

```python
edge_index = list(zip(edge_index[0], edge_index[1]))
corr_norm = list(zip(edge_index_normal[0], edge_index_normal[1]))
```

```python
corr_norm = set(corr_norm)
edge_index = set(edge_index)
edge_index_normal = corr_norm.intersection(edge_index)
```

```python
len(edge_index_normal)
```

```python
#BUILD EDGE INDEX FOR TUMOR
mask = correlation_matrix_tumor > threshold_edge

# Exclude the diagonal elements (assuming the matrix is square)
np.fill_diagonal(mask, False)

# Get the row and column indices where the mask is True
row_indices, col_indices = np.where(mask)

# Create the edge_index tensor
edge_index_tumor = np.array([row_indices, col_indices], dtype=np.int64)

print(edge_index_tumor.shape)
```

```python
corr_tum = list(zip(edge_index_tumor[0], edge_index_tumor[1]))
```

```python
corr_tum = set(corr_tum)
edge_index = set(edge_index)
edge_index_tumor = corr_tum.intersection(edge_index)
```

#WGCNA

```python
import numpy as np
import pandas as pd
import networkx as nx
import seaborn as sns
import matplotlib.pyplot as plt

# Create a synthetic gene expression dataset
np.random.seed(42)

# Calculate correlation matrix
correlation_matrix = pd.DataFrame(np.corrcoef(df_normal.transpose().values)) #df_normal.corr()
```

```python
# Create a weighted co-expression network
power = 6
adjacency_matrix = np.abs(correlation_matrix) ** power

# Convert adjacency matrix to pandas DataFrame and then to networkx graph
G = nx.from_pandas_adjacency(adjacency_matrix)

# Calculate eigenvector centrality
eigenvector_centrality = nx.eigenvector_centrality(G)

# Add eigenvector centrality as a node attribute
nx.set_node_attributes(G, eigenvector_centrality, 'eigenvector_centrality')

# Visualize the network with node color based on eigenvector centrality
pos = nx.spring_layout(G)  # You can try different layout algorithms
```

```python
node_colors = list(eigenvector_centrality.values())
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=False, node_size=10, node_color=node_colors, cmap=plt.cm.Blues, edge_color='gray')
plt.title("Co-expression Network with Eigenvector Centrality")
plt.colorbar(label="Eigenvector Centrality")
plt.show()
```

#Hierarchical Clustering

```python
links = pd.read_csv('/content/drive/MyDrive/lung/link_index.csv').drop('Unnamed: 0', axis=1)
set_index = list(set(links['index']))
edge_index = np.array([links['index'], links['index1']])
edges = list(zip(edge_index[0], edge_index[1]))
```

```python
import networkx as nx
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

G = nx.Graph(edges)
adjacency_matrix = 0.5 + 0.5 * np.corrcoef(df_normal.transpose())
power = 6
adjacency_matrix = np.nan_to_num(adjacency_matrix ** power)
adjacency_matrix.shape
```

```python
# Calculate the TOM matrix
def calculate_tom(adjacency_matrix):
    node_degrees = np.sum(adjacency_matrix, axis=1)
    sqrt_degrees = np.sqrt(np.outer(node_degrees, node_degrees))
    similarity_matrix = adjacency_matrix / sqrt_degrees
    tom_matrix = np.maximum(0, similarity_matrix)
    return tom_matrix

tom_matrix = np.nan_to_num(calculate_tom(adjacency_matrix))
tom_matrix.shape
```

```python
# Perform hierarchical clustering
linkage_matrix = linkage(tom_matrix, method='average')
print(linkage_matrix.shape)
# Plot the dendrogram
dendrogram(linkage_matrix, labels=list(G.nodes()))
plt.show()
```
