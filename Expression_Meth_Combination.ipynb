#Combination meth and genes

```python
import torch

comb_normal = torch.zeros((df_normal.shape[0], df_normal.shape[1], 2))
df_normal = df_normal.reindex(sorted(df_normal.columns), axis=1)
comb_normal[:, :, 0] = torch.tensor(df_normal.values)
comb_normal.shape
```

```python
cols_normal = set(df_meth_normal.columns.values)
for i, g in enumerate(df_normal.columns.values):
  if g in links_meth['gene_id'].values:
    meth = links_meth[links_meth['gene_id'] == g]['Name'].values
    c = list(set(meth).intersection(cols_normal))
    mean = 1 - df_meth_normal.loc[:, c].mean(axis=1, skipna=True).fillna(value=1)
    comb_normal[:, i, 1] = torch.tensor(mean.values)
comb_normal
```

```python
comb_tumor = torch.zeros((df_tumor.shape[0], df_tumor.shape[1], 2))
df_tumor = df_tumor.reindex(sorted(df_tumor.columns), axis=1)
comb_tumor[:, :, 0] = torch.tensor(df_tumor.values)
comb_tumor.shape
```

```python
cols_tumor = set(df_meth_tumor.columns.values)
for i, g in enumerate(df_tumor.columns.values):
  if g in links_meth['gene_id'].values:
    meth = links_meth[links_meth['gene_id'] == g]['Name'].values
    c = list(set(meth).intersection(cols_tumor))
    mean = 1 - df_meth_tumor.loc[:, c].mean(axis=1, skipna=True).fillna(value=1)
    comb_tumor[:, i, 1] = torch.tensor(mean.values)
comb_tumor
```

```python
X = torch.cat((comb_normal, comb_tumor), dim=0)
print(X.shape)
N = X.shape[0]
perm_indices = torch.randperm(N)
X = X[perm_indices] #shuffle
```

```python
y = torch.cat((torch.zeros(comb_normal.shape[0]), torch.ones(comb_tumor.shape[0])))
y = y[perm_indices]
y.shape
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
import torch
from torch.nn import Linear
from torch.nn import Conv1d
from torch.nn import Dropout

features_in = X_train.shape[2]
num_genes = X_train.shape[1]
n_classes = len(y_train.unique())
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class NET(torch.nn.Module):
    def __init__(self):
        super().__init__()
        torch.manual_seed(1234)
        self.conv1 = Conv1d(features_in, 1, kernel_size=3, stride=1, padding='same')
        #self.conv2 = Conv1d(16, 32, kernel_size=3, stride=1, padding=1)

        # Fully connected layers to make the final prediction
        self.fc1 = Linear(1 * num_genes, 1)
        #self.fc2 = Linear(1024, 512)
        #self.fc3 = Linear(512, 128)
        #self.fc4 = Linear(128, 1)

        self.dropout = Dropout(0.5)

    def forward(self, x):
        x = torch.relu(self.conv1(x))
        #x = torch.relu(self.conv2(x))

        # Reshape the tensor before passing it through fully connected layers
        x = x.view(x.size(0), -1)
        #out = x

        # Fully connected layers with dropout
        x = self.fc1(x)
        """
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = torch.relu(self.fc2(x))
        x = self.dropout(x)
        x = torch.relu(self.fc3(x))
        x = self.dropout(x)
        x = self.fc4(x)
        """

        # Sigmoid activation for binary classification (tumor or not)
        x = torch.sigmoid(x)

        return x#, out
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model = NET().to(device)
#model.load_state_dict(torch.load('/content/drive/MyDrive/BioInfo/Louvain_combined/weights_combination_5000.pth'))
model
```

```python
#TRAIN SETTINGS
import torch.optim as optim
import torch.nn as nn

loss_fn = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

n_epochs = 4000
```

```python
x_train = X_train.permute(0,2,1).to(device)
y_train = y_train.to(device)
```

```python
import numpy as np

#TRAIN LOOP
for epoch in range(n_epochs):
    optimizer.zero_grad()
    out = model(x_train).view(-1)
    loss = loss_fn(out, y_train)
    # backward pass
    loss.backward()
    if epoch % 100 == 0:
      print(f"EPOCH {epoch}, LOSS: {loss}")
    # update weights
    optimizer.step()
```

```python
x_test = X_test.permute(0,2,1).to(device)
y_test = y_test.to(device)
```

```python
#TEST
model.eval()
out = model(x_test).view(-1)
#pred = out.argmax(dim=1)  #Use the class with highest probability.
#print(out)
#print(y_test)
test_correct = out == y_test
test_acc = int(test_correct.sum()) / len(test_correct)
print("Accuracy on test set: {}".format(test_acc))
```
